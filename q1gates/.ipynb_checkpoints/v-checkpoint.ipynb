# V

__Matrix representation__   
$
V = \frac{1}{2} \begin{bmatrix} 1+i & 1-i \\ 1-i & 1+i \end{bmatrix}, \qquad V^\dagger = \frac{1}{2} \begin{bmatrix} 1-i & 1+i \\ 1+i & 1-i \end{bmatrix}
$

We use $V$ since $SX$ can be mistaken as the product of $S$ and $X$.

__Actions__
1. `V` changes the `Z`-basis to the `Y`-basis with additional phase:  
$\qquad V|0\rangle = e^{i\pi/4}|R\rangle, \quad V|1\rangle = e^{-i\pi/4} |L\rangle$  
$\qquad V^\dagger|0\rangle = e^{-i\pi/4}|L\rangle, \quad V^\dagger|1\rangle = e^{i\pi/4} |R\rangle$.   

2. `V` does not change the `X`-basis up to the phase factor:  
$\qquad V|+\rangle = |+\rangle, \quad V|-\rangle = i|-\rangle$  
$\qquad V^\dagger|+\rangle = |+\rangle, \quad V^\dagger|-\rangle = -i|-\rangle$.

3. `V` changes the `Y`-basis to the `Z`-basis with additional phase:  
$\qquad V|L\rangle = e^{i\pi/4}|0\rangle, \quad V|R\rangle = e^{-i\pi/4} | 1\rangle$  
$\qquad V^\dagger |L\rangle = e^{i\pi/4}|1\rangle, \quad V^\dagger|R\rangle = e^{-i\pi/4} |0\rangle$.

__Equivalent gates__

1. $V\, V = V^\dagger \, V^\dagger = X, \quad \Rightarrow \quad V = X V^\dagger = V^\dagger X, \quad V^\dagger = X V = V X$.
1. $V = H S H \quad \Rightarrow \quad H V = S H, \quad V H = H S$   
 $V^\dagger = H S^\dagger H \quad \Rightarrow \quad H V^\dagger = S^\dagger H, \quad V^\dagger H = H S^\dagger$
2. $V \simeq R_x\left(\frac{\pi}{2}\right), \qquad V^\dagger \simeq R_x\left(-\frac{\pi}{2}\right)$.

__Remarks__

1. Qiskit uses gate name `sx` for $V$ and shows it as $\sqrt{X}$ in chircuit diagrams.
2. Qiskit uses gate name `sxdg` for $V^\dagger$ and shows it as $\sqrt{Xdg}$ in chircuit diagrams.

## Checking the properties of $V$  with Qiskit quantum circuits

```python
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector, Operator
from qiskit.visualization import array_to_latex, plot_bloch_multivector, plot_state_qsphere
```

```python
# Get the matrix representation of V.
qr = QuantumRegister(1)
qc = QuantumCircuit(qr)
# Add V to the circuit as sx.
qc.sx(0)
array_to_latex(Operator(qc))
```

```python
# Get the matrix representation of V^\dagger.
qr = QuantumRegister(1)
qc = QuantumCircuit(qr)
# Add V to the circuit as sx.
qc.sxdg(0)
array_to_latex(Operator(qc))
```

---
Actions on X, Y, and Z-basis

```python
qr = QuantumRegister(3)
qc = QuantumCircuit(qr)
# prepare the 1st qubit in |1>, the 2nd  in |+> , and the 3rd in |L>.
qc.x(0)
qc.h(1)
qc.rx(-np.pi/2,2)
qc.draw()
```

```python
# Display the current state
psi=Statevector(qc)
print("Initial state:")
plot_bloch_multivector(psi)
```

```python
# Apply V to each qubit
qc.barrier()
qc.sx([0,1,2])
qc.draw()
```

```python
# Show that the Bloch vectors rotates around the $x$ axis.
psi=Statevector(qc)
plot_bloch_multivector(psi)
```

---
Equivalent gates

```python
qr = QuantumRegister(4)
qc = QuantumCircuit(qr)
# Flip qubits 1 and 3
qc.x([1,3])
```

```python
# Display the current state
psi=Statevector(qc)
print("Initial state:")
plot_bloch_multivector(psi)
```

```python
# Apply V on qubit 0 and 1, HSH on qubits 2 and 3
qc.barrier()
qc.sx([0,1])
qc.h([2,3])
qc.s([2,3])
qc.h([2,3])
qc.draw()
```

```python
# Show that V and HSH rotate the Bloch vectors exactly in the same way.
psi=Statevector(qc)
print("After V")
plot_bloch_multivector(psi)
```
